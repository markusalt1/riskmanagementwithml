```python
import joblib
import matplotlib.pyplot as plt
import numpy as np

#folder where estimators produced with and without importance sampling are saved
filepath = ".../Resultate_final/Put/saved_single_run/"
```

```python
#load risk measure estimators produced without IS
VaR_hat_NN = joblib.load(filepath+'VaR_hat_NN.joblib')
ES_hat_NN = joblib.load(filepath+'ES_hat_NN.joblib')
GlueVaR_hat_NN = joblib.load(filepath+'GlueVaR_hat_NN.joblib')
VaR_hat_RF = joblib.load(filepath+'VaR_hat_RF.joblib')
ES_hat_RF = joblib.load(filepath+'ES_hat_RF.joblib')
GlueVaR_hat_RF = joblib.load(filepath+'GlueVaR_hat_RF.joblib')

#load risk mesure estimators produced with IS and M_{IS}=500,000
VaR_hat_IS_NN = joblib.load(filepath+'VaR_hat_IS_1_500_NN.joblib')
ES_hat_IS_NN = joblib.load(filepath+'ES_hat_IS_1_500_NN.joblib')
GlueVaR_hat_IS_NN = joblib.load(filepath+'GlueVaR_hat_IS_1_500_NN.joblib')
VaR_hat_IS_RF = joblib.load(filepath+'VaR_hat_IS_1_500_RF.joblib')
ES_hat_IS_RF = joblib.load(filepath+'ES_hat_IS_1_500_RF.joblib')
GlueVaR_hat_IS_RF = joblib.load(filepath+'GlueVaR_hat_IS_1_500_RF.joblib')
```

```python
step_number = 200
i = 500000/step_number
steps = np.linspace(i, 500000, step_number)
```

Comparison plots that show estimators produced with the neural network or the random forest approach and their counterparts produced with IS in the same plot.

```python
plt.figure(figsize=(6.5,5),dpi=250)
plt.plot(steps, np.full(len(steps),8.3356), color='tab:green')
plt.plot(steps, VaR_hat_NN, color='tab:blue')
plt.plot(steps, VaR_hat_IS_NN, color='tab:purple')
plt.legend(['reference', 'neural network', 'IS neural network'])
plt.show()
```

```python
plt.figure(figsize=(6.5,5),dpi=250)
plt.plot(steps, np.full(len(steps),8.3356), color='tab:green')
plt.plot(steps, VaR_hat_RF, color='tab:orange')
plt.plot(steps, VaR_hat_IS_RF, color='tab:red')
plt.legend(['reference', 'random forest', 'IS random forest'])
plt.show()
```

```python
plt.figure(figsize=(6.5,5),dpi=250)
plt.plot(steps, np.full(len(steps),8.509), color='tab:green')
plt.plot(steps, ES_hat_NN, color='tab:blue')
plt.plot(steps, ES_hat_IS_NN, color='tab:purple')
plt.legend(['reference', 'neural network', 'IS neural network'])
plt.show()
```

```python
plt.figure(figsize=(6.5,5),dpi=250)
plt.plot(steps, np.full(len(steps),8.509), color='tab:green')
plt.plot(steps, ES_hat_RF, color='tab:orange')
plt.plot(steps, ES_hat_IS_RF, color='tab:red')
plt.legend(['reference', 'random forest', 'IS random forest'])
plt.show()
```

```python
plt.figure(figsize=(6.5,5),dpi=250)
plt.plot(steps, np.full(len(steps),7.689), color='tab:green')
plt.plot(steps, GlueVaR_hat_NN, color='tab:blue')
plt.plot(steps, GlueVaR_hat_IS_NN, color='tab:purple')
plt.legend(['reference', 'neural network', 'IS neural network'])
plt.show()
```

```python
plt.figure(figsize=(6.5,5),dpi=250)
plt.plot(steps, np.full(len(steps),7.689), color='tab:green')
plt.plot(steps, GlueVaR_hat_RF, color='tab:orange')
plt.plot(steps, GlueVaR_hat_IS_RF, color='tab:red')
plt.legend(['reference', 'random forest', 'IS random forest'])
plt.show()
```
